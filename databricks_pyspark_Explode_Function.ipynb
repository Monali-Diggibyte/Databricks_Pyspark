```
array_appliacnces = [('Raja', ['Tv','oven', 'AC', 'Refrigerator']),
                     ('Raghav', ['Tv',None, 'AC', '']),
                     ('Ravindra', ['Tv','Grinder']),
                     ('Ramesh', ['AC', 'Refrigerator', None]),
                     ('Rajesh', None),
                    ]

df_app =  spark.createDataFrame(data= array_appliacnces, schema= ['Name', 'Electronic Appliances'] )
df_app.printSchema()
display(df_app)
```

```
map_brand = [('Raja', {'Tv':'LG','oven':'Phillips', 'AC':'Blue Star', 'Refrigerator':'Whirlpool'}),
             ('Raghav', {'Tv':'samsung', 'AC':'LG', '':''}),
             ('Ravindra', {'Tv':'Sony','Grinder':'IFB'}),
             ('Ramesh', {'AC':'Samsung', 'Refrigerator':'IFB'}),
             ('Rajesh', None),
            ]

df_brand = spark.createDataFrame(data= map_brand, schema= ['name', 'Brand'])
df_brand.printSchema()
display(df_brand)
```

```
from pyspark.sql.functions import explode

display(df_app)
df_app.printSchema()

df = df_app.select(df_app.Name, explode(df_app['Electronic Appliances']))
display(df)
df.printSchema()
```

```
display(df_brand)
df_brand.printSchema()
```

```
df_b = df_brand.select(df_brand.name, explode(df_brand.Brand))
display(df_b)
df_b.printSchema()
```

```
df_b = df_b.withColumnRenamed('key',"Appliances").withColumnRenamed('value','Brand')

display(df_b)
```

```
# previously row with all null values excluded so this function include all null value row as well
from pyspark.sql.functions import explode_outer

display(df_app.select(df_app.Name, explode_outer(df_app['Electronic Appliances'])))

display(df_brand.select(df_brand.name, explode_outer(df_brand.Brand)))
```

```
from pyspark.sql.functions import posexplode,posexplode_outer
```

```
display(df_app.select(df_app.Name, posexplode(df_app['Electronic Appliances'])))

display(df_app.select(df_app.Name, posexplode_outer(df_app['Electronic Appliances'])))
```
