```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```
# same like creating table in SQL
schema_custom = StructType([StructField('Employee_id', IntegerType(), True),
                            StructField('Emp_First_name', StringType(), True),
                            StructField('Emp_Last_Same', StringType(), True),
                            StructField('Experience', IntegerType(), True),
                            StructField('Salary', IntegerType(), True),
                           ])
```

```
# same like loading data to table from file : recommended
df_c = spark.read.format("csv").schema(schema_custom).options(header= True, sep= ",").load("/FileStore/tables/emp_details.csv")

display(df_c)
```

```
df_c.printSchema()
```

```
# alternate approach to create custom schema
schema_alternate = "Employee_id INTEGER, Emp_First_name STRING, Emp_Last_Same STRING, Experience INTEGER, Salary INTEGER"
```
