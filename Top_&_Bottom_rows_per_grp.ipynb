```
input_data = [('kamal','Physics', 80, 'P', 90),
              ('kamal','Chemistry', 67, 'P', 90),
              ('kamal','Mathematics', 78, 'P', 90),
              ('Vimal','Physics', 30, 'F', 80),
              ('Vimal','Chemistry', 59, 'P', 80),
              ('Vimal','Mathematics', 75, 'P', 80),
              ('Rekha','Physics', 90, 'P', 70),
              ('Rekha','Chemistry', 87, 'P', 70),
              ('Rekha','Mathematics', 97, 'P', 70),
              ('john','Physics', 33, 'F', 60),
              ('john','Chemistry', 28, 'F', 60),
              ('john','Mathematics', 52, 'P', 60),
              ('Akash','Physics', 89, 'P', 75),
              ('Akash','Chemistry', 76, 'P', 75),
              ('Akash','Mathematics', 63, 'P', 75)
              ]

input_schema = ["Name", "Subject", 'Mark', "Status", 'Attendance']

df_std = spark.createDataFrame(data= input_data, schema= input_schema)

df_std.display()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number
```

```
window_dept= Window.partitionBy('Subject').orderBy(col('Mark').desc())

df2= df_std.withColumn('row', row_number().over(window_dept)).orderBy("Name", "row")

display(df2)
```

```
df3 = df2.filter(col("row")<= 1)  ## Top N records/subject

df3.display()
```

```
window_dept= Window.partitionBy('Subject').orderBy(col('Mark').asc())

df4= df_std.withColumn('row', row_number().over(window_dept)).orderBy("Name", "row")

display(df4)
```

```
df5 = df4.filter(col("row")<= 1)  ## Bottom N records/subject

df5.display()
```

```
window_dept= Window.partitionBy('Name').orderBy(col('Mark').desc())

df6= df_std.withColumn('row', row_number().over(window_dept)).orderBy("Name", "row")

display(df6)

df7 = df6.filter(col("row")<= 1)  ## Top N records/name 

df7.display()
```

```
window_dept= Window.partitionBy('Name').orderBy(col('Mark').asc())

df8= df_std.withColumn('row', row_number().over(window_dept)).orderBy("Name", "row")

display(df8)

df9 = df8.filter(col("row")<= 1)  ## Top N records/name 

df9.display()
```
