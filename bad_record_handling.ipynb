```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([StructField("Month", StringType(), True), \
                     StructField("emp_count", IntegerType(), True), \
                     StructField("production_unit", IntegerType(), True), \
                     StructField("expense", IntegerType(), True), \
                     StructField("_Corrupt_Record", StringType(), True)
                    ])
```

```
#df_p= spark.read.format("csv").options(header= True, mode="PERMISSIVE").schema(schema).load("/FileStore/tables/Production_corrupt-1.csv")

df_p= spark.read.format("csv").option("mode", "PERMISSIVE").option("header", "true").schema(schema).load("/FileStore/tables/Production_corrupt-2.csv")

df_p.display()
```

```
df_dm= spark.read.format("csv").option("mode", "DROPMALFORMED").option("header", "true").schema(schema).load("/FileStore/tables/Production_corrupt-2.csv")

df_dm.display()
```
