```
emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data = [(10, "raj Kumar", "1999","100", "M", 2000),
            (20, "Ravi Kumar", "2002","200", "M", 3000),
            (30, "Arjun Kumar", "2004","400", "M", 25000),
            (40, "Rani Kumari", "2006","100", "F", 20000),
            (50, "Rekha", "2008","300", None , 9000),
            (60, "raj Kumar", "2010",None, None, 2000)
            ]

df_emp = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(df_emp)
```

```
dept_schema = ["Department_ID", "Department Name"]

dept_data = [(100, "HR"),
             (200, "Networking"),
             (300, "Accounts"),
             (400, "IT"),
             (500, "Supply"),
             (600, "Sales"),
             (700, "Marketing")
            ]

df_dept = spark.createDataFrame(data= dept_data, schema= dept_schema)

display(df_dept)
```

```
display(df_emp)
display(df_dept)
```

```
## df1.join(df2, df1.key == df2.key, "jointype")

df_inner_join = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID,"inner")

df_inner_join = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, )  # by default its inner join# 'inner', 'outer', 'full', 'fullouter', 'full_outer', 'leftouter', 'left', 'left_outer', 'rightouter', 'right', 'right_outer', 'leftsemi', 'left_semi', 'semi', 'leftanti', 'left_anti', 'anti', 'cross'

display(df_inner_join)
```

```
df_left_outer = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, "leftouter") ## 'leftouter', 'left', 'left_outer'

display(df_left_outer)

df_left_outer1 = df_dept.join(df_emp, df_dept.Department_ID == df_emp.Dept_Id, "left_outer")

display(df_left_outer1)
```

```
df_right_outer = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, "rightouter")  ##'rightouter', 'right', 'right_outer'
display(df_right_outer)

df_right_outer1 = df_dept.join(df_emp, df_dept.Department_ID == df_emp.Dept_Id, "right" )
display(df_right_outer1)

df_right_outer2 = df_dept.join(df_emp, df_emp.Dept_Id == df_dept.Department_ID, "right_outer" )
display(df_right_outer2)
```

```
df_full_outer = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, 'fullouter') ## 'full','fullouter', 'full_outer', 'outer'
display(df_full_outer)

df_full_outer1 = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, 'outer')
display(df_full_outer1)
```

```
display(df_emp)
display(df_dept)

df_left_semi = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, 'leftsemi') ## 'leftsemi', 'left_semi', 'semi'
display(df_left_semi)
```

```
df_left_anti = df_emp.join(df_dept, df_emp.Dept_Id == df_dept.Department_ID, 'leftanti') ## 'leftanti', 'left_anti', 'anti'
display(df_left_anti)
```
