```
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('MyPractise').getOrCreate()
spark
```

```
emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data = [(10, "raj Kumar", "1999","100", "M", 2000),
            (20, "Ravi Kumar", "2002","200", "M", 3000),
            (30, "Arjun Kumar", "2004","400", "M", 25000),
            (40, "Rani Kumari", "2006","100", "F", 20000),
            (50, "Rekha", "2008","300", None , 9000),
            (60, "raj Kumar", "2010",None, None, 2000)
            ]

df_emp = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(df_emp)
```

```
emp_schema1 = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data1 = [(70, "Stephen", "1999","100", "M", 6000),
            (80, "Martin", "2002","200", "M", 9000),
            (90, "Angelika", "2004","400", "F", 85000),
            (100, "Selina", "2006","100", None, 10000),
            (50, "Rekha", "2008","300", None , 9000),
            (60, "raj Kumar", "2010",None, None, 2000)
            ]

df_emp1 = spark.createDataFrame(data= emp_data1, schema= emp_schema1)

display(df_emp1)
```

```
df_union = df_emp.union(df_emp1)  # contains duplicates :giving all records from both Dfs
display(df_union)
```

```
display(df_union.dropDuplicates())
```
