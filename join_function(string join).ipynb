```
emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data = [(10, "raj Kumar", "1999-01-01","100", "M", 2000),
            (20, "Ravi Kumar", "2002-02-02","200", "M", 3000),
            (30, "Arjun Kumar", "2004-03-03","400", "M", 25000),
            (40, "Rani Kumari", "2006-04-04","100", "F", 20000),
            (50, "Rekha", "2008-05-18","300", None , 9000),
            (60, "raj Kumar B", None,None, None, 2000)
            ]

df_emp = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(df_emp)
```

```
emp_columns = df_emp.columns

print(emp_columns)
```

```
joined_columns = "**".join(emp_columns)

print(joined_columns)
```

```
join_string= " AND ".join(list(map(lambda x: ("table1." + x + " = table2."+ x), emp_columns)))

print(join_string)
```

```
update_string= " , ".join(list(map(lambda x: ("table1." + x + " = table2."+ x), emp_columns)))

print(update_string)
```

```
insert_string= " , ".join(list(map(lambda x: ("table1." + x ), emp_columns)))

print(insert_string)
```
