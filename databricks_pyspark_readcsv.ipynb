```
import pyspark
```

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('MyPractise').getOrCreate()
```

```
spark
```

```
#Reading single File
df_s = spark.read.format("csv").option("inferSchema", True).option("header", True).option("sep",",").load("/FileStore/tables/emp_details.csv")

display (df_s)
# gives number of records/row counts
print(df_s.count())
```

```
#Reading single File
df_s = spark.read.format("csv").option("inferSchema", True).option("header", False).option("sep",",").load("/FileStore/tables/emp_details.csv")

display (df_s)
# gives number of records/row counts
print(df_s.count())
```

```
#Reading single File
df_s = spark.read.format("csv").options(inferSchema= True, header= True, sep= ",").load("/FileStore/tables/emp_details.csv")

display (df_s)
# gives number of records/row counts
print(df_s.count())
```

```
#Reading Multiple files (combine 2 or more files into single dataframe)
df_m = spark.read.format("csv").options(inferSchema= True, header= True, sep= ",").load(["/FileStore/tables/emp_details-1.csv","/FileStore/tables/emp_details.csv"])
 
display (df_m)

print(df_m.count())
```

```
#Read all CSV files under a Folder
df_all = spark.read.format("csv").options(inferSchema= True, header= True, sep= ",").load("/FileStore/tables/")
 
display (df_all)

print(df_all.count())

type(df_all)
```
