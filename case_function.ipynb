```
Schema_student = ["name", "Subject", "marks", "Exam Status", "Attendance"]

data_student = [("Raja","Maths",80, "P", 90),
                ("Ramesh","Hindi",85, "P", 70),
                ("Ram","Marathi",65, "P", 90),
                ("Rekha","English",35, "F", 60),
                ("Robert","Science",None, "NA", 70)
               ]

df_student = spark.createDataFrame(data= data_student, schema= Schema_student)

df_student.show()
```

```
from pyspark.sql.functions import when
```

```
display(df_student)

df1 = df_student.withColumn("Exam Status", when(df_student.marks >= 50 , "Pass")
                                          .when(df_student.marks < 50, "Fail")
                                          .otherwise("Abscent")
                           )

display(df1)
```

```
df2 = df_student.withColumn("New_Status", when(df_student.marks >= 50 , "Pass")
                                          .when(df_student.marks < 50, "Fail")
                                          .otherwise("Abscent")
                           )
display(df2)
```

```
from pyspark.sql.functions import expr
```

```
df3 = df_student.withColumn("New_Status", expr("CASE when marks>= 50 then 'Pass'"+ 
                                                    "when marks < 50 then 'Fail'" + 
                                                    "ELSE 'Abscent' END"))
display(df3)
```

```
display(df_student)

from pyspark.sql.functions import when

df4 = df_student.withColumn("Grade", when((df_student.marks >= 75), "Distiction")
                                    .when((df_student.marks < 75) & (df_student.marks >= 60), 'First')
                                    .when((df_student.marks < 60) & (df_student.marks >= 35), "Second")
                                    .otherwise("Fail"))  ## use '&' for 'and', '|' for 'or', '~' for 'not' when building DataFrame boolean expressions

display(df4)
```

```
display(df_student)

from pyspark.sql.functions import when

df4 = df_student.withColumn("Grade", when((df_student.marks >= 75) | (df_student.Attendance >= 80), "Distiction")
                                    .when(((df_student.marks >= 60) & (df_student.marks < 75)) | (df_student.Attendance < 80), 'First but less attendance')
                                    .when((df_student.marks >= 35) & (df_student.marks < 60) | (df_student.Attendance < 80), "Second but less attendance")
                                    .otherwise("Fail"))  ## use '&' for 'and', '|' for 'or', '~' for 'not' when building DataFrame boolean expressions

display(df4)
```
