```
df_emp = spark.read.format("csv").options(header = True, sep =",", inferschema= True).load("/FileStore/tables/employee.csv")

display(df_emp)
```

```
print(df_emp.count())
```

```
print(df_emp.rdd.getNumPartitions())
```

```
from pyspark.sql.functions import spark_partition_id

df_emp.withColumn("Partition_ID", spark_partition_id()).groupby("Partition_ID").count().show()
```

```
display(df_emp)

df_5 = df_emp.select(df_emp.Employee_id, df_emp.Emp_First_name, df_emp.Emp_Last_Same, df_emp.Experience).repartition(5)

print(df_5.rdd.getNumPartitions())

```

```
df_5.withColumn("Partition_ID", spark_partition_id()).groupby("Partition_ID").count().show() ## distribute records amoung the partiotions evenly.
```
