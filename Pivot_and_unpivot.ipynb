```
data = [("ABC", "Q1", 2000),
        ("ABC", "Q2", 4000),
        ("ABC", "Q3", 6000),
        ("ABC", "Q4", 12000),
        ("Xyz", "Q1", 10000),
        ("Xyz", "Q2", 5000),
        ("Xyz", "Q3", 2500),
        ("Xyz", "Q4", 1250),
        ("KLM", "Q1", 2000),
        ("KLM", "Q2", 2000),
        ("KLM", "Q3", 2000),
        ("KLM", "Q4", 2000),
       ]

Schema = ["Company", "Quarter", "Profit"]

df = spark.createDataFrame(data= data, schema= Schema)

display(df)
```

```
data = [("ABC", "Q1", 2000),
        ("ABC", "Q1", 2020),
        ("ABC", "Q2", 4000),
        ("ABC", "Q3", 6000),
        ("ABC", "Q4", 12000),
        ("Xyz", "Q1", 10000),
        ("Xyz", "Q2", 5000),
        ("Xyz", "Q3", 2500),
        ("Xyz", "Q4", 1250),
        ("KLM", "Q1", 2000),
        ("KLM", "Q2", 2000),
        ("KLM", "Q3", 2000),
        ("KLM", "Q4", 2000),
       ]

Schema = ["Company", "Quarter", "Profit"]

df_q = spark.createDataFrame(data= data, schema= Schema)

display(df_q)
```

```
df_pivot = df.groupby("Company").pivot("Quarter").sum("Profit")

df_pivot.show()
```

```
df_pivot_q = df_q.groupby("Company").pivot("Quarter").sum("Profit")

df_pivot_q.show()
```

```
df_unpivot = df_pivot.selectExpr("Company", "stack(4,'Q1',Q1, 'Q2',Q2, 'Q3',Q3, 'Q4',Q4) as (Quarter, Revenue)")

display(df_unpivot)
```

```
df_unpivot_q = df_pivot_q.selectExpr("Company", "stack(4,'Q1',Q1, 'Q2',Q2, 'Q3',Q3, 'Q4',Q4) as (Quarter, Revenue)")

display(df_unpivot_q)
```
