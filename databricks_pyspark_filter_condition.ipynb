```
emp_data = [(10, "raj Kumar", "1999","100", "M", 2000),
            (20, "Ravi Kumar", "2002","200", "M", 3000),
            (30, "Arjun Kumar", "2004","400", "M", 25000),
            (40, "Rani Kumari", "2006","100", "F", 20000),
            (50, "Rekha", "2008","300", None , 9000),
            (60, "raj Kumar", "2010",None, None, 2000)
            ]

emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_df = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(emp_df)
```

```
display(emp_df)

display(emp_df.filter(emp_df.salary == 2000)) # similarly for ==, <, >, <=, >=, !=
```

```
# and => & filter
display(emp_df.filter((emp_df.Gender == 'F') & (emp_df.DOJ == 2000)))
```

```
# OR => | filter
display(emp_df.filter((emp_df.Gender == 'F') | (emp_df.DOJ == 2002)))
```

```
display(emp_df)
display(emp_df.filter((emp_df.Name.endswith("r")) & (emp_df.Name.startswith("R"))))
```

```
display(emp_df.filter((emp_df.Gender.isNull())))
```

```
display(emp_df.filter((emp_df.Gender.isNotNull())))
```

```
display(emp_df.filter((emp_df.Dept_Id.isNull())))
```

```
display(emp_df)

display(emp_df.filter((emp_df.Dept_Id.isin(100, 200))))
```

```
display(emp_df.filter((~emp_df.Dept_Id.isin(100, 200))))
```

```
display(emp_df.filter((emp_df.Name.like("%Kumar%"))))
```

```
display(emp_df)

display(emp_df.filter(~(emp_df.Name.like("%Kumar%"))))
```
