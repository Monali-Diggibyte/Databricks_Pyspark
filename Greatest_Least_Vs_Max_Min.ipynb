```
input_data = [('kamal', 70, 68,89, 40, 84),
              ('Vimal', 90, 67, 87, 79, 74),
              ('Rekha', 66, 88, 49, 65, 72),
              ('john', 78, 73, 82, 89, 67),
              ('Akash', 80, 86, 69, 78, 92)
              ]

input_schema= ['Student Name', 'Science', 'Mathematics', 'Social Studies', 'English', 'Hindi']

df_marks = spark.createDataFrame(data= input_data, schema= input_schema)

df_marks.display()
```

```
from pyspark.sql.functions import greatest

df_greatest = df_marks.withColumn('Greatest', greatest('Science', 'Mathematics', 'Social Studies', 'English', 'Hindi'))

df_greatest.display()
```

```
df_greatestof2 = df_marks.withColumn('Greatest', greatest('Science', 'Mathematics'))

df_greatestof2.display()
```

```
from pyspark.sql.functions import least

df_g_least= df_greatest.withColumn("Least", least('Science', 'Mathematics', 'Social Studies', 'English', 'Hindi'))

df_g_least.display()
```

```
df_marks.agg({"Science": 'Max',"Mathematics": 'max'}).display()
```

```
df_marks.agg({"Science": 'min', "Social Studies": "Min"}).display()
```

```
df_marks.agg({"Science": 'min',"Mathematics": 'max', "Social Studies": "Min", 'English': 'Max'}).display()
```
