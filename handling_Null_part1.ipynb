```
emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data = [(10, "raj Kumar", "1999-01-01","100", "M", 2000),
            (20, "Ravi Kumar", "2002-02-02","200", "M", 3000),
            (30, "Arjun Kumar", "2004-03-03","400", "M", 25000),
            (40, "Rani Kumari", "2006-04-04","100", "F", 20000),
            (50, "Rekha", "2008-05-18","300", None , 9000),
            (60, "raj Kumar B", None,None, None, 2000)
            ]

df_emp = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(df_emp)
```

```
display(df_emp.filter(df_emp.Gender.isNull()))
```

```

display( df_emp.filter("Gender IS NULL"))
```

```
from pyspark.sql.functions import col

display(df_emp.filter(col("Gender").isNull()))
```

```
#Syntax1:
display(df_emp.filter(df_emp.Gender.isNotNull()))

#Syntax2:
display(df_emp.filter("Gender IS NOT NULL"))

#Syntax3:
from pyspark.sql.functions import col
display(df_emp.filter(col("Gender").isNotNull()))
```
