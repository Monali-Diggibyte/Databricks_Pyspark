```
emp_schema = ["Employee_ID", "Name", "DOJ", "Dept_Id", "Gender", "salary"]

emp_data = [(10, "raj Kumar", "1999","100", "M", 2000),
            (20, "Ravi Kumar", "2002","200", "M", 3000),
            (30, "Arjun Kumar", "2004","400", "M", 25000),
            (40, "Rani Kumari", "2006","100", "F", 20000),
            (50, "Rekha", "2008","300", None , 9000),
            (60, "raj Kumar", "2010",None, None, 2000)
            ]

df_emp = spark.createDataFrame(data= emp_data, schema= emp_schema)

display(df_emp)
```

```
from pyspark.sql.functions import lit
```

```
emp_addcolumn = df_emp.withColumn("Location", lit("Mumbai")).show()
```

```
emp_addcolumn = df_emp.withColumn("Location", lit("Mumbai"))

display(emp_addcolumn)
```

```
df_emp_addcolumn1 = df_emp.withColumn("Bonus", df_emp.salary*.10)

type(df_emp)
display(df_emp_addcolumn1)
```

```
from pyspark.sql.functions import concat
```

```
display(df_emp)

df_emp_addcolumn2 = df_emp.withColumn("Bonus", df_emp.salary*.10).withColumn("New Dept_Id", concat("Employee_ID", lit("_"), "Dept_Id")) #concat(lit("1 "), "Dept_Id" # lit("1 "),

df_emp_addcolumn2.show()
```

```
df_emp_renameColmn = df_emp.withColumnRenamed("Name", "Full Name")

df_emp_renameColmn.show()
```
